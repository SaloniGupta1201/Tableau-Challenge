```python
# NYC Citibike Program 2020 - get 2020 January-December tripdata
import pandas as pd
import os
```

```python
# Get NYC Citibike tripdata csv files from 201806 to 201808
filepath_jan = os.path.join("Resources" ,"JC-202001-citibike-tripdata.csv")
filepath_feb = os.path.join("Resources" ,"JC-202002-citibike-tripdata.csv")
filepath_mar = os.path.join("Resources" ,"JC-202003-citibike-tripdata.csv")
filepath_apr = os.path.join("Resources" ,"JC-202004-citibike-tripdata.csv")
filepath_may = os.path.join("Resources" ,"JC-202005-citibike-tripdata.csv")
filepath_jun = os.path.join("Resources" ,"JC-202006-citibike-tripdata.csv")
filepath_jul = os.path.join("Resources" ,"JC-202007-citibike-tripdata.csv")
filepath_aug = os.path.join("Resources" ,"JC-202008-citibike-tripdata.csv")
filepath_sep = os.path.join("Resources" ,"JC-202009-citibike-tripdata.csv")
filepath_oct = os.path.join("Resources" ,"JC-202010-citibike-tripdata.csv")
filepath_nov = os.path.join("Resources" ,"JC-202011-citibike-tripdata.csv")
filepath_dec = os.path.join("Resources" ,"JC-202012-citibike-tripdata.csv")


nycbike_2020jan = pd.read_csv(filepath_jan)
nycbike_2020feb = pd.read_csv(filepath_feb)
nycbike_2020mar = pd.read_csv(filepath_mar)
nycbike_2020apr = pd.read_csv(filepath_apr)
nycbike_2020may = pd.read_csv(filepath_may)
nycbike_2020jun = pd.read_csv(filepath_jun)
nycbike_2020jul = pd.read_csv(filepath_jul)
nycbike_2020aug = pd.read_csv(filepath_aug)
nycbike_2020sep = pd.read_csv(filepath_sep)
nycbike_2020oct = pd.read_csv(filepath_oct)
nycbike_2020nov = pd.read_csv(filepath_nov)
nycbike_2020dec = pd.read_csv(filepath_dec)
```

```python
nycbike_jan_dec_df = pd.concat([nycbike_2020jan,nycbike_2020feb,nycbike_2020mar,nycbike_2020apr,nycbike_2020may,nycbike_2020jun,
nycbike_2020jul,nycbike_2020aug,nycbike_2020sep,nycbike_2020oct,nycbike_2020nov,nycbike_2020dec])
```

```python
nycbike_jan_dec_df.tail(10)
```

```python
# Check if there is any null value
nycbike_jan_dec_df.isna().sum()
```

```python
# Output concatenated 2018 July-Aug citibike tripdata to CSV file 
filepath_2020_jan_dec = os.path.join("Resources" ,"citibiketripdata.csv")
nycbike_jan_dec_df.to_csv(filepath_2020_jan_dec, index=False)
```
